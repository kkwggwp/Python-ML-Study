### 데이터 색인 (심화)
- 행과 열의 범위를 주어 데이터를 추출

##### 인덱서 (loc,iloc)
- df.loc[행범위,열범위] : 컬럼명, 인덱스명을 통해서 추출
- df.iloc[행범위,열범위]: 인덱스 번호를 통해서 추출 / 끝범위 +1

### 데이터 추가(행,열)

- 열(column) 추가
  - df['추가하고싶은 컬럼명'] = 데이터
  - 가장 마지막에 추가

- 열(column) 삽입
  - df.insert(loc = 인덱스, column = 컬럼명, value = 데이터)
  - 입력한 인덱스 자리에 삽입, 이후데이터는 뒤로 밀림

- 행(row) 추가
  - df.loc["추가하고싶은 인덱스"] = 데이터
  - column 즉 특성의 개수를 맞춰줘야한다!
